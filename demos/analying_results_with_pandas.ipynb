```python
import pandas as pd

df = pd.read_csv("data/plots/experiment.csv").fillna("")
df
```

```python
# lets just see the total result values.
# we use a df['level'] == 0 to create a filter and df[filter] to apply it.

df[df["level"] == 0]
```

```python
# if we wanted to select several specific nodes, we could use the isin() method (again as a filter).
df[df["node_name"].isin(["wind", "solar"])]
```

```python
# filters can be combined with | or &.
# so if we would have a filter like this:
df[df["results_GWP1000_magnitude"] > 0.2]
```

```python
# we could combine it with another filter like this:
df[df["node_name"].isin(["wind", "solar"]) & (df["results_GWP1000_magnitude"] > 0.1)]
```

```python
# filters can be stored in order to make the code more readable.
node_filter = df["node_name"].isin(["wind", "solar"])
gwp_filter = df["results_GWP1000_magnitude"] > 0.1
```

```python
df[gwp_filter & node_filter]
```

```python
df[df["level"] == 0].sort_values("output_0_magnitude", ascending=False)
```

```python
df[df["level"] == 0].sort_values(
    ["output_0_magnitude", "results_GWP1000_magnitude"], ascending=False
)
```

```python
# the result of all these operations is a new dataframe, which we can assign to a variable (in order to apply more operations later)
level_filter = df["level"] == 2
min_output_filter = df["output_0_magnitude"] >= 3
filtered_df = df[level_filter & min_output_filter]

filtered_df
```

```python
# this is how we drop (remove) columns from a dataframe.
remove_columns = [
    "level",
    "parent_name",
    "results_GWP1000_unit",
    "results_LandUse_unit",
    "results_WCP_unit",
    "output_0_unit",
    "output_0_label",
    "bw_activity_code",
]
filtered_df.drop(remove_columns, axis=1)
```
